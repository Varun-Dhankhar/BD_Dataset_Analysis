# One method to reduce the effect of CPD:

#to reduce effect if mean of current identified state from cpd and the next state is same upto 2nd decimal place than combine both of them into one state only. 

```python
import math
def truncate(f, n):
    n=2
    return math.floor(f * 10 ** n) / 10 ** n


mod_results=results23.copy()
lem=[]


for value in range(0, len(mod_results)-2):
    i=mod_results[value]
    j=mod_results[value+1]
    k=mod_results[value+2]
    df_copy_ij=df_d2.copy()
    df_copy_jk=df_d2.copy()
    start_time=df_d2.loc[i]['Timestamp']
    end_time_j=df_d2.loc[j]['Timestamp']
    end_time_k=df_d2.loc[k]['Timestamp']
    df_copy_ij=df_copy_ij[(df_copy_ij['Timestamp']>=start_time) & (df_copy_ij['Timestamp']<=end_time_j)]
    df_copy_jk=df_copy_jk[(df_copy_jk['Timestamp']>=end_time_j) & (df_copy_jk['Timestamp']<=end_time_k)]
    a=df_copy_ij['rolling_average_minutely'].mean()
    b=df_copy_jk['rolling_average_minutely'].mean()
    a=truncate(a, 2)
    b=truncate(b, 2)
    if a==b or abs(a-b)<0.02:
        lem.append(j)
        
        
```
