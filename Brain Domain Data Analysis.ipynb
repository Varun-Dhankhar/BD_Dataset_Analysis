```python
# Import pandas
import pandas as pd
import csv
 
# reading csv file
df2=pd.read_csv("D:/Time-varyying_PF_data/06-December-2022.csv")

%matplotlib notebook
```

```python
header=["Timestamp", 'V1','V2', 'V3', 'A1', 'A2', 'A3', 'PA1', 'PA2', 'PA3', 'W1','W2', 'W3', 'VAR1', 'VAR2', 'VAR3', 'VA1', 'VA2', 'VA3', 'F', 'A12', 'A13', 'PF', 'Date', 'Time', 'DWh', 'DVARh', 'RWh', 'RVARh', 'TW', 'TVAR', 'TVA', 'TA', 'V12', 'V23', 'V31', 'THD1', 'THD2', 'THD3', 'THDA1', 'THDA2', 'THDA3', 'PF1', 'PF2', 'PF3']

df2.columns=header
```

# Preprocessing

```python
dfp2=df2.copy()
```

```python
import datetime


dfp2['Time'] = pd.to_datetime(df2['Timestamp'], unit='s', utc=False)
#dfp= dfp['Meter_Time'].dt.strftime('%d/%m/%Y %H:%M:%S')
#dfp['Date'] = pd.to_datetime(df1.Meter_Time, unit='s')
#dfp['Date']=dfp.Date.dt.strftime('%d/%m/%Y %H:%M:%S')

dfp2['Time']=dfp2['Time'].dt.tz_localize('UTC').dt.tz_convert('Asia/Singapore')

dfp2['Time']= dfp2['Time'].dt.strftime('%d/%m/%Y %H:%M:%S')
dfp2[['Date','Time']]=dfp2['Time'].str.split(' ', 1, expand=True)
dfp2.columns
```

```python
#2
dfp2=dfp2.drop(columns=['Date'])
df8c2=dfp2['Time']
df9c2=dfp2['Timestamp']
dfp2=dfp2.drop(columns=['Timestamp', 'Time']).astype('float')
dfp2['Timestamp']=df9c2
dfp2['Time']=df8c2
```

```python
dfp2['minutes']=dfp2['Time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1])+ (1/60)*int(x[2]))
dfp2['minutes']
```

```python
df_2=dfp2[(dfp2['minutes']>=938) & (dfp2['minutes']<=963)] #either feed entire data or just the corresponding minutes data
df_2.plot(x='minutes', y=["A1"],  figsize=(10,4))
```

```python
%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt2
import pandas as pd
import numpy as np


data = df_2['demean-A1'].astype(float).values

N = data.shape[0] #number of elements
t = np.linspace(0, N) #converting minutes to seconds
s = data

fft = np.fft.fft(s)
fftfreq = np.fft.fftfreq(len(s))

T = t[1] - t[0]

f = np.linspace(0, 1 / T, N)
plt2.figure(figsize = (10, 6))
plt2.ylabel("Amplitude")
plt2.xlabel("Frequency [Hz]")
plt2.figure(1)
plt2.plot(fftfreq, np.abs(fft))
plt2.show()
```
